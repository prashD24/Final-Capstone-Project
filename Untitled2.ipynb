## Capstone Project – The Battle of Neighborhoods | Finding better localities in Toronto for placing the ATM's

### Introduction

The purpose of this Capstone Project is to help banks in understanding the socio geographic spread of Toronto across its neighborhood. It will help bank making smart and efficient decision on selecting great neighborhood out of numbers of other neighborhoods in Toranto.

Facility location is a problem of paramount importance and optimizing business operations without affecting customer service is very challenging. In the case of banking services, the location of bank branches and ATMs must match the service demands (turn around time for service, reachability etc) of the customers' and the expected quality of service is determined by the socio-economic background of the customer. Therefore, it is necessary to formulate the optimization problem so as to reflect the customers' expectations and tolerance for quality of service in a given geographical region.

The ability to do so requires clustering people living in the region into several smaller areas based on postal codes. An ideal clustering algorithm should consider the social behavior of people living in the service area and the uncertainty associated with their social behavior.

### Methodology

The raw dataset is prepared by combining the data from all the above three sources. A dataframe is prepared with Population and count of venues across ten primary catgeories as provided by foursquare. The box plot is drawn for all these 11 paramters.  


Most venues are falling under the category of Food, Shops and Service and Professional & Other places. Events and Residences have very little data and hence they can be dropped from the dimension list for clustering.

The raw dataset is normalized using the standard scaler function of the sklearn library to prepare for ingestion by K-Means clustering algorithm. 

K-Means clustering is used for clustering the dataset into similar clusters. iterating through various cluster numbers, I settle for the 4 clusters as it provides the optimal results for 4 clusters.
